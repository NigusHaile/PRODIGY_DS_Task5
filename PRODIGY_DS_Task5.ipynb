```python
# Plotting a Pandas.Series data
missing_Accident_data = Accidents.isna().sum().sort_values(ascending=False) * 100. / len(Accidents)
```

```python
missing_Accident_data.plot(kind='bar', color='skyblue', title='Accident Data', ylabel='Values')
fig = plt.figure(figsize=(100,5))
plt.show()
```

```python
Accidents.City.unique()
```

```python
cities_by_accident = Accidents.City.value_counts()
cities_by_accident[0:9]
```

```python
cities_by_accident[0:9].plot(kind='bar', color='blue', title='Accident Data', ylabel='Values')
fig = plt.figure(figsize=(5,100))
plt.show()
```

```python
import seaborn as sns
sns.histplot(cities_by_accident, bins=25, kde=True)
plt.title('cities_by_accident')
plt.show()
```

```python
high_accident_cities = cities_by_accident[cities_by_accident >=1000] # having over 1000 accidents
low_accident_cities = cities_by_accident[cities_by_accident < 1000]  # having less than 1000 accidents
```

```python
# Distribution of high accident cities
sns.displot(high_accident_cities)
```

```python
# Distribution of high accident cities
sns.histplot(low_accident_cities)
```

```python
# Distribution of low accident cities
sns.histplot(low_accident_cities, log_scale=True)
```
